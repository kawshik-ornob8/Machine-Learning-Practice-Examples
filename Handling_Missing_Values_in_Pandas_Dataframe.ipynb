#How to Handle Missing Values in Pandas

### Filling Null Values

We use fillna to replace null value to some value like

```python
df2 = df.fillna(value = 0)
df2
```

```python
#Filling Null Values with a Previous value

df3 = df.fillna(method = 'pad')
df3
```

```python
df3.isnull().sum()
```

```python
#Filling Null Values with the Next value

df4 = df.fillna(method = 'bfill')
df4
```

```python
#Filling Null Values with a Previous Columns value
df5 = df.fillna(method = 'pad' , axis = 1)
df5
```

```python
#Filling Null Values with a next Columns value
df6 = df.fillna(method = 'bfill' , axis = 1)
df6
```

```python
#Filling Different Values in Null in different Columns

df7 = df.fillna({'society': 'abcd', 'balcony': 'efgh'})
df7
```

```python
#Filling Null VAlue with the mean of a column

df8 = df.fillna(value=df['balcony'].mean())
df8
```

```python
#Filling Null VAlue with the max of a column

df9 = df.fillna(value=df['balcony'].max())
df9
```

```python
#Filling Null VAlue with the min of a column

df10 = df.fillna(value=df['balcony'].min())
df10
```

#Dropna()

```python
df11 = df.dropna()
df11
```

```python
#if the Columns Contain all null value
df12 = df.dropna(how='all')
df12
```

```python
#if the Columns Contain any null value
df13 = df.dropna(how='any')
df13
```

#Replace

```python
import numpy as np
df14 = df.replace(to_replace=np.nan, value =1525)
df14
```

```python
df15 = df.replace(to_replace=3.0, value =5.0)
df15
```
