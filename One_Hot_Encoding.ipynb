```python
#Create a object of OneHotEncoder

ohe = OneHotEncoder()
```

```python
#Now applyed fit and transfrom
ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
#Now Save this result in a variable
feature_array = ohe.fit_transform(df[['Colour','Country']]).toarray()
```

```python
#to see the categories
ohe.categories_
```

```python
feature_labels = ohe.categories_
```

You see here categories store in 2 defferent array...
To save one single array we can use ***ravel()***

```python
np.array(feature_labels).ravel()
```

```python
feature_labels = np.array(feature_labels).ravel()
```

```python
#Now print varible
pd.DataFrame(feature_array, columns = feature_labels)
```

```python
features = pd.DataFrame(feature_array, columns = feature_labels)
features
```

```python
features.head()
```

```python
#to join with main dataset we use concat
update_df = pd.concat([df,features],axis=1)
update_df
```
